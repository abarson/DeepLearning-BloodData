```python
import matplotlib.pyplot as plt
import csv
```

```python
fname = "lrcn-training.log"

accs = []
losses = []
val_accs = []
val_losses = []
with open(fname,'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    header = next(reader)
    
    for row in reader:
        acc, loss, val_acc, val_loss = map(float,row[1:])
        accs.append(round(acc,4))
        losses.append(round(loss,4))
        val_accs.append(round(val_acc,4))
        val_losses.append(round(val_loss,4))

print(len(accs))
```

```python
fig, axarr = plt.subplots(2,1,sharex=True)

plt.subplots_adjust(hspace=0.5)
plt.suptitle("Results",fontsize=14)
#axarr[0].set_title("Accuracy",fontsize=14)
axarr[0].set_ylabel("Accuracy",fontsize=14)
axarr[0].plot(accs,label='training accuracy')
axarr[0].plot(val_accs,label='validation accuracy')
#axarr[1].set_title("Loss",fontsize=14)
axarr[1].plot(losses,label='training')
axarr[1].plot(val_losses,label='validation')
axarr[1].set_xlabel("Epoch",fontsize=14)
axarr[1].set_ylabel("Loss",fontsize=14)

plt.legend(bbox_to_anchor=(0.65,1.5))
plt.savefig("figs/midway_results.pdf")
plt.show()
```
