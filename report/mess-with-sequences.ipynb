```python
import pandas as pd
import os
import sys
```

```python
frame_dir = "../frames/"
csv_ = "../NextStartingPoint.csv"
DL_class_data = "../DeepLearningClassData.csv"
```

```python
nxt_start_point = pd.read_csv(csv_)
dldf = pd.read_csv(DL_class_data)
nxt_start_point
```

```python
subjects = [s for s in os.listdir(frame_dir)]
frame_df = pd.DataFrame(columns=["Subject", "Trial", "Heart Rate", "Respiratory Rate"])

columns = dldf.columns
print(columns)
i = 0
for index, row in nxt_start_point.iterrows():
    subject, trial = row['Subject'], row['Trial']
    
    data = dldf[dldf['SUBJECT'] == subject]
    
    subj, t1_hrate, t1_resprate, t2_hrate, t2_resprate = [list(data[col])[0] for col in columns]
    if int(trial) == 1:
        row = [subject, trial, t1_hrate, t1_resprate]
    
    else:
        row = [int(subject), int(trial), t2_hrate, t2_resprate]
    
    frame_df.loc[i] = row
    i+=1
    
frame_df
```
